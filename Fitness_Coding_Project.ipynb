```python
#importing necessary library for the kNN algorithm
import numpy
```

```python
#make predictions on calories burned based on the Average speed of the person during their run, Average heart rate as they run, as well as the
#distance they ran
def knn(sample, data, k):
  updated_fitness_data["kNN Distance"] = numpy.sqrt(
      (sample["Average Speed (km/h)"] - data["Average Speed (km/h)"])**2
      + (sample["Average Heart Rate (bpm)"] - data["Average Heart Rate (bpm)"])**2
      + (sample["Distance (km)"] - data["Distance (km)"])**2
  )
  sorted_data = data.sort_values(["kNN Distance"])

  #print(sorted_data.iloc[0:k][['Type', 'Average Speed (km/h)', 'Average Heart Rate (bpm)', 'Distance (km)', 'Calories Burned', 'kNN Distance']])
  #The above line of code can be uncommented to have an in-depth look as to how the kNN algorithm works
  return sorted_data.iloc[0:k]["Calories Burned"].mean()
```

```python
#normalizing the data in preparation for the kNN algorithm
speed_mean = updated_fitness_data["Average Speed (km/h)"].mean()
speed_standard_deviation = updated_fitness_data["Average Speed (km/h)"].std()

beat_mean = updated_fitness_data["Average Heart Rate (bpm)"].mean()
beat_standard_deviation = updated_fitness_data["Average Heart Rate (bpm)"].std()

distance_mean = updated_fitness_data["Distance (km)"].mean()
distance_standard_deviation = updated_fitness_data["Distance (km)"].std()

updated_fitness_data["Average Speed (km/h)"] = (updated_fitness_data["Average Speed (km/h)"] - speed_mean)/speed_standard_deviation
updated_fitness_data["Average Heart Rate (bpm)"] = (updated_fitness_data["Average Heart Rate (bpm)"] - beat_mean)/beat_standard_deviation
updated_fitness_data["Distance (km)"] = (updated_fitness_data["Distance (km)"] - distance_mean)/distance_standard_deviation
```

```python
#testing the kNN algorithm with normalization
fitness_sample = {}
fitness_sample["Average Speed (km/h)"] = 9.86
fitness_sample["Average Heart Rate (bpm)"] = 155.0
fitness_sample["Distance (km)"] = 12.94

knn(fitness_sample, updated_fitness_data, 4)
```

```python
#removing kNN Distance column since the function knn is done being called
updated_fitness_data = updated_fitness_data.drop("kNN Distance", axis=1)
updated_fitness_data
```
